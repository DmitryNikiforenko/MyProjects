### Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall

 Precision - метрика оценки точности предсказания, т. е. как много из предсказанных объектов относятся к интересующему классу.
 Максимизация этой метрики необходима в случаях макисмальной точности: распознование лиц в охранных системах, предсказание      генетического заболевания.
 
 Recall - метрика оценки полноты предсказания, т. е. как много объектов интересующего класса было предсказано.
 Эта метрика исспользуется для увеличение полноты охвата интересующего класса: любые скрининговые системы - диагностика раковых заболеваний путём оценки медицинских изображений RG, РКТ, МРТ; стратификация риска сердечно-сосудистых заболеваний.

### Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

F-мера позволяет объёдинить в одном показателе точность и полноту. В качестве результата она выдаёт их гармоничное среднее. Посредством изменения коэфициента мы можем акцентировать внимание на той метрике, которая для нас более приоритетна.

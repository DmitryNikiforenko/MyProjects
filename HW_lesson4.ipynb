### Расскажите, как работает регуляризация в решающих деревьях, какие параметры мы штрафуем в данных алгоритмах?

* Регуляризация вводит штраф за большие веса - метод борьбы с переобучением.
* L2 - регуляризация - наиболее частый выбор
* L1 - регуляризация - сложнее оптимизировать, но можно отбирать признаки. При исспользовании этого метода регуляризации признаки с нулевыми весами являются бесполезными и их можно отбросить.

### По какому принципу рассчитывается "важность признака (feature_importance)" в ансамблях деревьев?

* Это техника присваивания очков полезности зависимым переменным – Предикторам (Predictor Variables) в зависимости от того, насколько они способны спрогнозировать Целевую переменную (Target Variable).
* Стратегии на основе дерева, используемые случайными лесами, естественным образом ранжируются по тому, насколько хорошо они улучшают чистоту модели в плане данных. Таким образом, “обрезая” деревья ниже определенного коэффициента, мы можем подобрать наиболее важные признаки.

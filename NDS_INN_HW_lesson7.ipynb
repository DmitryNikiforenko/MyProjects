### Для примера рассматриваемой модели выберем YOLOv2

Основные тезисы, отражающие отличия YOLOv2 от YOLO:
* Добавили Batch Normalization. Как раз появилась оригинальная статья про батчнормализацию, и ее добавили во все сети, что совершенно логично при том насколько повышается скорость тренировки и качество результата
* В YOLOv1 сначала тренировался классификатор на картинках 224×224, потом брались натренированные веса, добавлялись полносвязные слои, вход расширялся до 448×448 и тренировался детектор. Теперь авторы предлагают после тренировки классификатора на малом разрешении, вначале потренировать его еще 10 эпох на повышенном разрешении, и только потом уже переходить к тренировки детектора
* Авторы заменили полносвязные слои свёрточными и перешли к схеме с якорями. Т.е. фактически к схеме SSD 
* Были модифицированы размеры прямоугольников якорей под датасет. Запустили k-mean кластеризацию на размерах прямоугольников объектов в тренировочном датасете и получили размеры, которые лучше для начальных предсказаний, чем как в SSD использовать просто разные соотношения сторон и масштабы.
* Изменили штрафную функцию на позиционирование, приблизив её к варианту, который используется в SSD и FasterRCNN
* Вместо того, чтобы предсказывать объекты только по последнему слою особенностей, добавили в предсказателя еще и предыдущий слой. Т.е. для входного изображения 448×448 на последнем слое было 13×13 особенностей, добавили с предыдущего слоя еще 28×28 таким образом добирая мелкие объекты - конкатенация с более ранними слоями для обнаружения более мелких деталей и сочетания их с с более обобщёнными признаками
* Поскольку сетка не привязана к конкретному размеру изображения, то при тренировке, каждые 10 эпох меняли случайным образом размер входного изображения выбирая в пределах от 320×320 до 608×608 с шагом 32. Фактически это похоже на тот способ расширения данных, который используется при тренировки SSD сети
